<h2 style="font-family:Helvetica New; text-align: left;"> Merge all the datasets in one big dataset of the year 2023: </h2> 

```r
trips_2023 <- bind_rows(trips_Dec23, trips_Nov23, trips_Oct23, trips_Sep23, 
                        trips_Aug23, trips_Jul23, trips_Jun23, trips_May23,
                        trips_Apr23, trips_Mar23, trips_Feb23, trips_Jan23)
```

<p style="font-size:17px; font-family:Helvetica New; line-height: 1.8em">   After merging all datasets, there is a need to clean and organise data. Next step will be to <strong>create some useful columns such as ride_duration, ride_month, ride_day_of_the_week, ride_distance</strong>:
</p>

<h2 style="font-family:Helvetica New; text-align: left;"> Create new helpful columns: </h2> 

```r
# Create ride duration column (in minutes):
trips_2023$ride_duration <- round(((as.double(difftime(trips_2023$ended_at, 
                                  trips_2023$started_at))) /60), digits=2)

# Creating ride distance column(in meters):
trips_2023$ride_distance <- round((distGeo(matrix(c(trips_2023$start_lng, 
                                                    trips_2023$start_lat), ncol = 2),
                                           matrix(c(trips_2023$end_lng, 
                                                    trips_2023$end_lat), ncol = 2))), 
                                  digits = 2)

# Creating ride month column:
trips_2023$ride_month <- format(as.Date(trips_2023$started_at), "%b")

# Creating ride day of the week column:
trips_2023$ride_day_week <- format(as.Date(trips_2023$started_at), "%A")

# Creating ride hour column:
trips_2023$ride_hour <- format(as.POSIXct(trips_2023$started_at),"%H")
```

<h2 style="font-family:Helvetica New; text-align: left;"> Cleaning Data </h2> 

```r
# Drop rows with not available or missing values (NA) and 
# Create new dataset for cleaned data:
trips_2023_cleaned <- drop_na(trips_2023)
glimpse(trips_2023_cleaned)

# Remove rows where ride distance is less than 100 m: 
trips_2023_cleaned <- trips_2023_cleaned[trips_2023_cleaned$ride_distance > 100,]

# Remove rows where ride duration is 0 or negative:
trips_2023_cleaned <- trips_2023_cleaned[trips_2023_cleaned$ride_duration > 0,]

# Remove the outliers where ride duration is more than 24 hours (1440 minutes):
trips_2023_cleaned <- trips_2023_cleaned[trips_2023_cleaned$ride_duration < 1440,]
```

<h1 style="font-family:Helvetica New; text-align: left;"> Analysis </h1>

<h2 style="font-family:Helvetica New; text-align: left;"> Bike type trends </h2> 

```r
# Create new dataset for analysis:
rides <- trips_2023_cleaned
glimpse(rides)

# Members vs Casual riders trend:
rider_type_count <- rides %>%
  group_by(member_casual) %>%
  reframe(total = n()) %>%
  mutate(overall_total = sum(total)) %>%
  group_by(member_casual) %>%
  reframe(percent_total = total/overall_total)

write.csv(rider_type_count, file = 'rider_type_count.csv')

# Bike type trends:
bike_type_count <- rides %>%
  group_by(rideable_type) %>%
  reframe(total = n()) %>%
  mutate(overall_total = sum(total)) %>%
  group_by(rideable_type) %>%
  reframe(percent_total = total/overall_total)

write.csv(bike_type_count, file = 'bike_type_count.csv')

# Bike types used by casual riders:
casual_bike_type_count <- rides %>%
  filter(member_casual == "casual") %>%
  group_by(rideable_type) %>%
  reframe(total = n()) %>%
  mutate(overall_total = sum(total)) %>%
  group_by(rideable_type) %>%
  reframe(percent_total = total/overall_total)

write.csv(casual_bike_type_count, file = 'casual_bike_type_count.csv')

# Bike types used by members:
member_bike_type_count <- rides %>%
  filter(member_casual == "member") %>%
  group_by(rideable_type) %>%
  reframe(total = n()) %>%
  mutate(overall_total = sum(total)) %>%
  group_by(rideable_type) %>%
  reframe(percent_total = total/overall_total)

write.csv(member_bike_type_count, file = 'member_bike_type_count.csv')
```

<h2 style="font-family:Helvetica New; text-align: left;"> Average amount: </h2> 

```r
# Average time per user type:
avrg_duration_user_type <- rides %>%
  group_by(member_casual) %>%
  reframe(avrg_duration = mean(ride_duration))
avrg_duration_user_type$avrg_duration <- round(avrg_duration_user_type$avrg_duration, 
                                               digits = 1)

write.csv(avrg_duration_user_type, file = 'avrg_duration_user_type.csv')

# Average distance per user type:
avrg_distance_user_type <- rides %>%
  group_by(member_casual) %>%
  reframe(avrg_distance = mean(ride_distance))
avrg_distance_user_type$avrg_distance <- round(avrg_distance_user_type$avrg_distance,
                                               digits = 0) 

write.csv(avrg_distance_user_type, file = 'avrg_distance_user_type.csv')

# Average distance per user type and byke type:
avrg_distance_user_bike <- rides %>%
  group_by(member_casual, rideable_type) %>%
  reframe(avrg_distance = mean(ride_distance))

write.csv(avrg_distance_user_bike, file = 'avrg_distance_user_bike.csv')

# Average duration per user type and byke type:
avrg_duration_user_bike <- rides %>%
  group_by(member_casual, rideable_type) %>%
  reframe(avrg_duration = mean(ride_duration))

write.csv(avrg_duration_user_bike, file = 'avrg_duration_user_bike.csv')
```

<h2 style="font-family:Helvetica New; text-align: left;"> Time and Seasonality </h2> 

```r
# Rides by hour of the day and by user type:
rides_by_hour <- rides %>%
  group_by(member_casual, ride_hour) %>%
  reframe(total = n())

write.csv(rides_by_hour, file = 'rides_by_hour.csv')

# Rides by month and by user type:
rides_by_month <- rides %>%
  group_by(member_casual, ride_month) %>%
  reframe(total = n()) 

rides_by_month$ride_month <- ordered(rides_by_month$ride_month, levels = c("Jan", "Feb", 
                                                                           "Mar", "Apr", "May",
                                                                           "Jun", "Jul", "Aug",
                                                                           "Sep", "Oct", "Nov",
                                                                           "Dec"))
write.csv(rides_by_month, file = 'rides_by_month.csv')

# Rides by day of the week and by user type:
rides_by_day <- rides %>%
  group_by(member_casual, ride_day_week) %>%
  reframe(total = n())

rides_by_day$ride_day_week <- ordered(rides_by_day$ride_day_week, levels = c("Monday", "Tuesday", 
                             "Wednesday", "Thursday", 
                             "Friday","Saturday", "Sunday"))

write.csv(rides_by_day, file = 'rides_by_day.csv')
```

<h2 style="font-family:Helvetica New; text-align: left;"> Most popular stations </h2> 

```r
# Rides by start station and by casual user:
start_station_casual <- rides %>%
  filter(member_casual == "casual") %>%
  group_by(start_station_name, member_casual) %>%
  reframe(total = n())

# Top 10 start stations most popular for casual riders:
start_station_casual_top10 <- top_n(start_station_casual, 10, start_station_casual$total) %>%
                                arrange(desc(total))

# Rides by start station by member:
start_station_member <- rides %>%
  filter(member_casual == "member") %>%
  group_by(start_station_name, member_casual) %>%
  reframe(total = n())

# Find the total members rides that correspond to top 10 stations for casual riders:
df <- start_station_member %>%
  filter(start_station_name %in% start_station_casual_top10$start_station_name)

start_station_casual_top10 = merge(y = df, x = start_station_casual_top10, 
                                   by = c("total", "start_station_name", "member_casual"), 
                                   all = TRUE)

write.csv(start_station_casual_top10, file = 'start_station_casual_top10.csv')

# Top 10 start stations most popular for members:
start_station_member_top10 <- top_n(start_station_member, 10, 
                                    start_station_member$total) %>%
                                    arrange(desc(total))

# Find the total casual rides that correspond to top 10 stations for members:
df <- start_station_casual %>%
  filter(start_station_name %in% start_station_member_top10$start_station_name)

start_station_member_top10 = merge(y = df, x = start_station_member_top10, 
                                   by = c("total", "start_station_name", "member_casual"), 
                                   all = TRUE)

write.csv(start_station_member_top10, file = 'start_station_member_top10.csv')

# Rides by end station and by casual user:
end_station_casual <- rides %>%
  filter(member_casual == "casual") %>%
  group_by(end_station_name, member_casual) %>%
  reframe(total = n())

# Top 10 end stations most popular for casual riders:
end_station_casual_top10 <- top_n(end_station_casual, 10, end_station_casual$total) %>% 
                              arrange(desc(total))

# Rides by end station by member:
end_station_member <- rides %>%
  filter(member_casual == "member") %>%
  group_by(end_station_name, member_casual) %>%
  reframe(total = n())

# Find the total members rides that correspond to top 10 stations for casual riders:
df <- end_station_member %>%
  filter(end_station_name %in% end_station_casual_top10$end_station_name)

end_station_casual_top10 = merge(y = df, x = end_station_casual_top10, 
                                 by = c("total", "end_station_name", "member_casual"), 
                                 all = TRUE)

write.csv(end_station_casual_top10, file = 'end_station_casual_top10.csv')

# Top 10 end stations most popular for members:
end_station_member_top10 <- top_n(end_station_member, 10, end_station_member$total) %>% 
                                  arrange(desc(total))

# Find the total casual rides that correspond to top 10 stations for members:
df <- end_station_casual %>%
  filter(end_station_name %in% end_station_member_top10$end_station_name)

end_station_member_top10 = merge(y = df, x = end_station_member_top10,  
                                 by = c("total", "end_station_name", "member_casual"), 
                                 all = TRUE)

write.csv(end_station_member_top10, file = 'end_station_member_top10.csv')
```
